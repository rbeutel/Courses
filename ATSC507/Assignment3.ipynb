# Assignment 3 - Coordinate Transformations

```python
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import math
```

```python
data = pd.read_csv('./n_hem.csv')
data.head()
```

### lat/lon coatline plot

```python
fig, ax = plt.subplots(1,1,figsize=(15,8))
ax.plot(data.lon_degr, data.lat_degr,lw=0.8)
ax.set_ylabel('Latitude [$^\circ$]', fontsize=14)
ax.set_xlabel('Longitude [$^\circ$]', fontsize=14);
```

### polar stereographic coastline plot

```python
plt.figure(figsize=(12, 12))
ax = plt.axes(projection=ccrs.Stereographic(60,0))

ax.plot(data.lon_degr, data.lat_degr,
             transform=ccrs.PlateCarree()) # why in the workd you need to use a platecarree projection in this step to get this to work i have no clue!
# ax.coastlines()
ax.set_xticks([0])
ax.set_yticks([0])
ax.set_title("Polar Stereographic, centered on 60$^\circ$N", fontsize=16)
```

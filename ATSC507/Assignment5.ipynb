## part B.5

```python
cr= np.linspace(0,3,100)
```

```python
L =  [2,  2.5,  3,  4,  5,  10,  20]
kx = np.array([2*math.pi/l for l in L])
```

```python
def RK3(cr, kx):
    A = np.sqrt(((cr**6)/36)*(math.sin(kx)**6)-((cr**4)/12)*(math.sin(kx)**4)+1)
    
    return A
```

```python
fig, ax = plt.subplots(1,1, figsize=(7,9))

for i in range(len(kx)):
    A = RK3(cr, kx[i])
    ax.plot(cr,A,label='L = {}$\Delta$x'.format(L[i]))
ax.legend()
ax.set_xlim([cr[0],cr[-1]])
ax.set_ylim([0.9,2.5])
```

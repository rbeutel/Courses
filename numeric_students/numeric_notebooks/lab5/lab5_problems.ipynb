```python
class Integ51(Integrator):
    def set_yinit(self):
        #
        # read in 'albedo_white chi S0 L albedo_black R albedo_ground'
        #
        uservars = namedtuple('uservars', self.config['uservars'].keys())
        self.uservars = uservars(**self.config['uservars'])
        #
        # read in 'whiteconc blackconc'
        #
        initvars = namedtuple('initvars', self.config['initvars'].keys())
        self.initvars = initvars(**self.config['initvars'])
        self.yinit = np.array(
            [self.initvars.whiteconc, self.initvars.blackconc])
        self.nvars = len(self.yinit)
        return None
    
    
    def __init__(self, coeffFileName):
        super().__init__(coeffFileName)
        self.set_yinit()

    def derivs5(self, y, t):
        """y[0]=fraction white daisies = 0.2
           y[1]=fraction black daisies = 0.7
           
           Constant growty rates for white
           and black daisies beta_w and beta_b
           
           returns dy/dt
        """
        user = self.uservars
        #
        # bare ground
        #
        x = 1.0 - y[0] - y[1]

#         growth rates don't depend on temperature
        beta_b = 0.7  # growth rate for black daisies
        beta_w = 0.7  # growth rate for white daisies

        # create a 1 x 2 element vector to hold the derivitive
        f = np.empty([self.nvars], 'float')
        f[0] = y[0] * (beta_w * x - user.chi)
        f[1] = y[1] * (beta_b * x - user.chi)
        return f


theSolver = Integ51('fixed_growth.yaml')
timeVals, yVals, errorList = theSolver.timeloop5fixed()

plt.close('all')
thefig, theAx = plt.subplots(1, 1)
theLines = theAx.plot(timeVals, yVals)
theLines[0].set_marker('+')
theLines[1].set_linestyle('--')
theLines[1].set_color('k')
theLines[1].set_marker('*')
theAx.set_title('constant growth rate')
theAx.set_xlabel('time')
theAx.set_ylabel('fractional coverage')
theAx.legend(theLines, ('white daisies', 'black daisies'), loc='best')

# thefig, theAx = plt.subplots(1, 1)
# theLines = theAx.plot(timeVals, errorList)
# theLines[0].set_marker('+')
# theLines[1].set_linestyle('--')
# theLines[1].set_color('k')
# theLines[1].set_marker('*')
# theAx.set_title('lab 5 interactive 1 errors')
# theAx.set_xlabel('time')
# theAx.set_ylabel('error')
# out = theAx.legend(theLines, ('white errors', 'black errors'), loc='best')
```

### Can you determine when non-zero steady states are achieved?

Yes, non-zero steady states will be achieved when there is no more space for the daisy population to grow at its specified rate. The daisies cannot grow by a fraciton of their growth rate so the populations will be capped, then at evey timesetp the daisies will oscilate between just dying or dying and growing. If the populations start out too high then the dasies will die by the death rate at every timestep until their is space for grwoth, at which point they will begin to oscillate (reached a steady state). If there are too few daisies at the beginning they will grow until there is no more space, die in the next timestep, and begin to oscilate. If the daidies have the same growth and death rate they will reach SS at the same time. <br>

For example let $\beta_w = \beta_b = 0.7$, $A_w = 0.2$ and $A_b = 0.6$, where A represents the fraction of surface area, and let the number of interger timesteps be represented by t, then you can use the following function to solve for how many timesteps it takes to reach the planet's limitting capacity:

```python
Aw = 0.3
Ab = 0.2
Chi = 0.2
Bw = 0.7 - Chi
Bb = 0.7 - Chi

Aw_test = Aw*(1+Bw)
Ab_test = Ab*(1+Bb)

if Aw_test+Ab_test >1:
    for t in range(50):
        need = Aw*Bw + Ab*Bb
        have = 1-(Aw+Ab)
        if need > have:
            Aw = (1-Chi)*Aw
            Ab = (1-Chi)*Ab
        if need < have:
            print(Aw)
            print(Ab)
            print(t)
            break
else:
    for t in range(50):
        need = Aw*Bw + Ab*Bb
        have = 1-(Ab+Aw)
        if have > need:
            Aw = (1+Bw)*Aw
            Ab = (1+Bb)*Ab
        if need > have:
            print(Aw)
            print(Ab)
            print(t)
            break
```

# Problem 2 - Coupling
### For the current value of L (0.2) in the file coupling.yaml, the final daisy steady state is zero. Why is it zero?
Not enough incident light to support daisy life - as shown bellow the more light the merrier. Begin to see life at around L=0.6.

### Find a value of L which leads to a non-zero steady state.

```python
class IntegCoupling(Integrator):
    """rewrite the init and derivs5 methods to
       work with a single (grey) daisy
    """
    def set_yinit(self, newL):
        # change the luminocity
        #
        self.config["uservars"]["L"] = newL # change solar incidence fraction
        #
        # read in 'albedo_grey chi S0 L  R albedo_ground'
        #
        uservars = namedtuple('uservars', self.config['uservars'].keys())
        self.uservars = uservars(**self.config['uservars'])
        #
        # read in 'greyconc'
        #
        initvars = namedtuple('initvars', self.config['initvars'].keys())
        self.initvars = initvars(**self.config['initvars'])
        self.yinit = np.array([self.initvars.greyconc])
        self.nvars = len(self.yinit)
        return None

    def __init__(self, coeffFileName, newL):
        super().__init__(coeffFileName)
        self.set_yinit(newL)

    def derivs5(self, y, t):
        """
           Make the growth rate depend on the ground temperature
           using the quadratic function of temperature

           y[0]=fraction grey daisies
           t = time
           returns f[0] = dy/dt
        """
        sigma = 5.67e-8  # Stefan Boltzman constant W/m^2/K^4
        user = self.uservars
        x = 1.0 - y[0]
        albedo_p = x * user.albedo_ground + y[0] * user.albedo_grey
        Te_4 = user.S0 / 4.0 * user.L * (1.0 - albedo_p) / sigma
        eta = user.R * user.L * user.S0 / (4.0 * sigma)
        temp_y = (eta * (albedo_p - user.albedo_grey) + Te_4)**0.25
        if (temp_y >= 277.5 and temp_y <= 312.5):
            beta_y = 1.0 - 0.003265 * (295.0 - temp_y)**2.0
        else:
            beta_y = 0.0

        # create a 1 x 1 element vector to hold the derivative
        f = np.empty([self.nvars], np.float64)
        f[0] = y[0] * (beta_y * x - user.chi)
        return f
```

```python
Lvals = np.linspace(0, 1, 11)


thefig, theAx = plt.subplots(1, 1, figsize=(8, 6))

for newL in Lvals:
    theSolver = IntegCoupling("coupling.yaml", newL)
    timeVals, yVals, errorList = theSolver.timeloop5fixed()
    
    if yVals[-1] > 0.01:
        print(theSolver.uservars.L)

    theLines = theAx.plot(timeVals, yVals, label = 'grey daisies, L='+str(theSolver.uservars.L))
    theAx.set_title('lab 5: interactive 2 Coupling with grey daisies') # \nSS = '+str(yVals[-1]))
    theAx.set_xlabel('time')
    theAx.set_ylabel('fractional coverage')
    out = theAx.legend(loc='best')
```

In this problem the daisy albedo is equal to the planet albedo so there is it not effecting planet temperature.

```python
from numlabs.lab5.lab5_funs import Integrator

class Integ54(Integrator):
    def set_yinit(self, newW, newB):
        # change initial fractions
        #
        self.config["initvars"]["whiteconc"] = newW # change fraction of white daisies
        self.config["initvars"]["blackconc"] = newB # change fraction of white daisies
        #
        #
        # read in 'albedo_white chi S0 L albedo_black R albedo_ground'
        #
        uservars = namedtuple('uservars', self.config['uservars'].keys())
        self.uservars = uservars(**self.config['uservars'])
        #
        # read in 'whiteconc blackconc'
        #
        initvars = namedtuple('initvars', self.config['initvars'].keys())
        self.initvars = initvars(**self.config['initvars'])
        self.yinit = np.array(
            [self.initvars.whiteconc, self.initvars.blackconc])
        self.nvars = len(self.yinit)
        return None

    def __init__(self, coeff_file_name, newW, newB):
        super().__init__(coeff_file_name)
        self.set_yinit(newW, newB)

    def find_temp(self, yvals):
        """
            Calculate the temperatures over the white and black daisies
            and the planetary equilibrium temperature given the daisy fractions
            
            input:  yvals -- array of dimension [2] with the white [0] and black [1]
                    daisy fractiion
            output:  white temperature (K), black temperature (K), equilibrium temperature (K)
        """
        sigma = 5.67e-8  # Stefan Boltzman constant W/m^2/K^4
        user = self.uservars
        bare = 1.0 - yvals[0] - yvals[1]
        albedo_p = bare * user.albedo_ground + \
            yvals[0] * user.albedo_white + yvals[1] * user.albedo_black
        Te_4 = user.S0 / 4.0 * user.L * (1.0 - albedo_p) / sigma
        temp_e = Te_4**0.25
        eta = user.R * user.L * user.S0 / (4.0 * sigma)
        temp_b = (eta * (albedo_p - user.albedo_black) + Te_4)**0.25
        temp_w = (eta * (albedo_p - user.albedo_white) + Te_4)**0.25
        return (temp_w, temp_b, temp_e)

    def derivs5(self, y, t):
        """y[0]=fraction white daisies
           y[1]=fraction black daisies
           no feedback between daisies and
           albedo_p (set to ground albedo)
        """
        temp_w, temp_b, temp_e = self.find_temp(y)

        if (temp_b >= 277.5 and temp_b <= 312.5):
            beta_b = 1.0 - 0.003265 * (295.0 - temp_b)**2.0
        else:
            beta_b = 0.0

        if (temp_w >= 277.5 and temp_w <= 312.5):
            beta_w = 1.0 - 0.003265 * (295.0 - temp_w)**2.0
        else:
            beta_w = 0.0
        user = self.uservars
        bare = 1.0 - y[0] - y[1]
        # create a 1 x 2 element vector to hold the derivitive
        f = np.empty_like(y)
        f[0] = y[0] * (beta_w * bare - user.chi)
        f[1] = y[1] * (beta_b * bare - user.chi)
        return f
```

# Problem 5 - Temperature

### Override timeloop5fixed so that it saves these three temperatures, plus the daisy growth rates to new variables in the Integ54 instance

```python
class Integ54(Integrator):
    def set_yinit(self, newW, newB):
        # change initial fractions
        #
        self.config["initvars"]["whiteconc"] = newW # change fraction of white daisies
        self.config["initvars"]["blackconc"] = newB # change fraction of white daisies
        #
        #
        # read in 'albedo_white chi S0 L albedo_black R albedo_ground'
        #
        uservars = namedtuple('uservars', self.config['uservars'].keys())
        self.uservars = uservars(**self.config['uservars'])
        #
        # read in 'whiteconc blackconc'
        #
        initvars = namedtuple('initvars', self.config['initvars'].keys())
        self.initvars = initvars(**self.config['initvars'])
        self.yinit = np.array(
            [self.initvars.whiteconc, self.initvars.blackconc])
        self.nvars = len(self.yinit)
        return None

    def __init__(self, coeff_file_name, newW, newB):
        super().__init__(coeff_file_name)
        self.set_yinit(newW, newB)

    def find_temp(self, yvals):
        """
            Calculate the temperatures over the white and black daisies
            and the planetary equilibrium temperature given the daisy fractions
            
            input:  yvals -- array of dimension [2] with the white [0] and black [1]
                    daisy fractiion
            output:  white temperature (K), black temperature (K), equilibrium temperature (K)
        """
        sigma = 5.67e-8  # Stefan Boltzman constant W/m^2/K^4
        user = self.uservars
        bare = 1.0 - yvals[0] - yvals[1]
        albedo_p = bare * user.albedo_ground + \
            yvals[0] * user.albedo_white + yvals[1] * user.albedo_black
        Te_4 = user.S0 / 4.0 * user.L * (1.0 - albedo_p) / sigma
        temp_e = Te_4**0.25
        eta = user.R * user.L * user.S0 / (4.0 * sigma)
        temp_b = (eta * (albedo_p - user.albedo_black) + Te_4)**0.25
        temp_w = (eta * (albedo_p - user.albedo_white) + Te_4)**0.25
        return (temp_w, temp_b, temp_e)

    def derivs5(self, y, t):
        """y[0]=fraction white daisies
           y[1]=fraction black daisies
           no feedback between daisies and
           albedo_p (set to ground albedo)
        """
        temp_w, temp_b, temp_e = self.find_temp(y)

        if (temp_b >= 277.5 and temp_b <= 312.5):
            beta_b = 1.0 - 0.003265 * (295.0 - temp_b)**2.0
        else:
            beta_b = 0.0

        if (temp_w >= 277.5 and temp_w <= 312.5):
            beta_w = 1.0 - 0.003265 * (295.0 - temp_w)**2.0
        else:
            beta_w = 0.0
        user = self.uservars
        bare = 1.0 - y[0] - y[1]
        # create a 1 x 2 element vector to hold the derivitive
        f = np.empty_like(y)
        f[0] = y[0] * (beta_w * bare - user.chi)
        f[1] = y[1] * (beta_b * bare - user.chi)
        return f, temp_w, temp_b, temp_e, beta_w, beta_b
    
    def rkckODE5(self, yold, timeStep, deltaT):

        # initialize the Cash-Karp coefficients
        # defined in the tableau in lab 4,

        a, c1, c2, b = self.rkckConsts
        i = self.initvars
        # set up array to hold k values in lab4 
        derivArray = np.empty([6, self.nvars], 'float')
        ynext = np.zeros_like(yold)
        bsum = np.zeros_like(yold)
        estError = np.zeros_like(yold)
        # vector k1 in lab4 equation 3.9
        derivArray[0, :], temp_w, temp_b, temp_e, beta_w, beta_b = self.derivs5(yold, timeStep)[:]

        # calculate step
        # c1=c_i in lab 4 notation, but c2=c_i - c^*_i

        y = yold
        for i in np.arange(5):
            bsum = 0.
            for j in np.arange(i + 1):
                bsum = bsum + b[i, j] * derivArray[j, :]
            # vectors k2 through k6 in lab4 
            # pdb.set_trace()
            derivArray[i + 1, :], temp_w, temp_b, temp_e, beta_w, beta_b= self.derivs5(y + deltaT * bsum, timeStep + a[i] * deltaT)[:]
            # partial sum of error in lab4 \Delta_est
            #
            #  sum the error term
            #
            estError = estError + c2[i] * derivArray[i, :]
            # print "estError: ",estError
            #
            # 5th order estimate y_{n+1}
            #
            ynext = ynext + c1[i] * derivArray[i, :]
        # final fifth order anser
        y = y + deltaT * (ynext + c1[5] * derivArray[5, :])
        # final 4th order estimate estimate
        estError = deltaT * (estError + c2[5] * derivArray[5, :])
        # print "estError final: ",estError
        timeStep = timeStep + deltaT
#       pdb.set_trace()
        return (y, estError, timeStep, temp_w, temp_b, temp_e, beta_w, beta_b)

    def timeloop5fixed(self):
        """fixed time step with
           estimated errors
        """
        t = self.timevars
        yold = self.yinit
        yError = np.zeros_like(yold)
        yvals = [yold]
        errorList = [yError]
        temp_eList = [yError]
        temp_wList = [yError]
        temp_bList = [yError]
        beta_wList = [yError]
        beta_bList = [yError]
        timeSteps = np.arange(t.tstart, t.tend, t.dt)
        for theTime in timeSteps[:-1]:
            yold, yError, newTime, temp_w, temp_b, temp_e, beta_w, beta_b = self.rkckODE5(yold, theTime, t.dt)
#             _, temp_w, temp_b, temp_e = self.deriv5(yold, theTime)
            #pdb.set_trace()
            yvals.append(yold)
            errorList.append(yError)
            temp_eList.append(temp_e)
            temp_bList.append(temp_b)
            temp_wList.append(temp_w)
            beta_bList.append(beta_b)
            beta_wList.append(beta_w)
        yvals = np.array(yvals).squeeze()
        errorVals = np.array(errorList).squeeze()
        return (timeSteps, yvals, errorVals, temp_wList, temp_bList, temp_eList, beta_wList, beta_bList)
```

### Play with the time step and final time, attempting small changes at first. How reasonable is the error estimate?

```python
from numlabs.lab5.lab5_funs import Integrator


class Integ55(Integrator):
    def set_yinit(self, new_t, new_dt):
        # change initial t and dt
        #
        self.config["timevars"]["tend"] = new_t 
        self.config["timevars"]["dt"] = new_dt 
        #
        # read in 'c1 c2 c3'
        #
        uservars = namedtuple('uservars', self.config['uservars'].keys())
        self.uservars = uservars(**self.config['uservars'])
        #
        # read in initial yinit
        #
        initvars = namedtuple('initvars', self.config['initvars'].keys())
        self.initvars = initvars(**self.config['initvars'])
        self.yinit = np.array([self.initvars.yinit])
        self.nvars = len(self.yinit)
        return None

    def __init__(self, coeff_file_name, new_t, new_dt):
        super().__init__(coeff_file_name)
        self.set_yinit(new_t, new_dt)

    def derivs5(self, y, theTime):
        """
           y[0]=fraction white daisies
        """
        user = self.uservars
        f = np.empty_like(self.yinit)
        f[0] = user.c1 * y[0] + user.c2 * theTime + user.c3
        return f
```

```python
theSolver = Integ55('expon.yaml')

# adaptive stepsize

timeVals, yVals, yErrors = theSolver.timeloop5Err()
timeVals = np.array(timeVals)
exact = timeVals + np.exp(-timeVals)
yVals = np.array(yVals)
yVals = yVals.squeeze()
yErrors = np.array(yErrors)

thefig, theAx = plt.subplots(1, 2, figsize = (14,4))
realestError = yVals - exact
actualErrorLine = theAx[0].plot(timeVals, realestError, label='actual error')
estimatedErrorLine = theAx[0].plot(timeVals, yErrors, label='estimated error')
theAx[0].legend(loc='best')
theAx[0].set_title('adaptive errors')

# fixed stepsize

timeVals, yVals, yErrors = theSolver.timeloop5fixed()
np_yVals = np.array(yVals).squeeze()
yErrors = np.array(yErrors)
np_exact = timeVals + np.exp(-timeVals)

realestError = np_yVals - np_exact
actualErrorLine = theAx[1].plot(timeVals, realestError, label='actual error')
estimatedErrorLine = theAx[1].plot(timeVals, yErrors, label='estimated error')
theAx[1].legend(loc='best')
theAx[1].set_title('fixed errors')
```

```python
import matplotlib.pyplot as plt
import context
from numlabs.lab2.lab2_functions import euler,beuler,leapfrog
import numpy as np
```

```python
theFuncs={'euler':euler,'beuler':beuler,'leapfrog':leapfrog}
```

```python
theLambda=-8.
stradle= -2/theLambda
print("at a lambda of "+str(theLambda)+" the stability condition is a delta t of "+str(stradle))
```

```python
def stability(npts, y1, y2, y3):
    tend=10.
    Ta=20.
    To=30.
    #npts=40
    
    fig, axs = plt.subplots(1, 3, figsize=(20, 6))
    
    funChoice='euler'
    approxTime,approxTemp=theFuncs[funChoice](npts,tend,To,Ta,theLambda)
    exactTime=np.empty([npts,],float)
    exactTemp=np.empty_like(exactTime)
    for i in np.arange(0,npts):
       exactTime[i] = tend*i/npts
       exactTemp[i] = Ta + (To-Ta)*np.exp(theLambda*exactTime[i])
    
    axs[0].plot(exactTime,exactTemp,'r-+')
    axs[0].plot(approxTime,approxTemp)
    axs[0].set_xlim([0,10])
    axs[0].set_ylim([10,y1])
    axs[0].set_title('euler, t= '+str(tend/npts))

    funChoice='leapfrog'
    approxTime,approxTemp=theFuncs[funChoice](npts,tend,To,Ta,theLambda)
    exactTime=np.empty([npts,],float)
    exactTemp=np.empty_like(exactTime)
    for i in np.arange(0,npts):
       exactTime[i] = tend*i/npts
       exactTemp[i] = Ta + (To-Ta)*np.exp(theLambda*exactTime[i])
    
    axs[1].plot(exactTime,exactTemp,'r-+')
    axs[1].plot(approxTime,approxTemp)
    axs[1].set_xlim([0,10])
    axs[1].set_ylim([-10000000,y2])
    axs[1].set_title('leapfrog, t= '+str(tend/npts))
    
    funChoice='beuler'
    approxTime,approxTemp=theFuncs[funChoice](npts,tend,To,Ta,theLambda)
    exactTime=np.empty([npts,],float)
    exactTemp=np.empty_like(exactTime)
    for i in np.arange(0,npts):
       exactTime[i] = tend*i/npts
       exactTemp[i] = Ta + (To-Ta)*np.exp(theLambda*exactTime[i])
    
    axs[2].plot(exactTime,exactTemp,'r-+')
    axs[2].plot(approxTime,approxTemp)
    axs[2].set_xlim([0,10])
    axs[2].set_ylim([10,y3])
    axs[2].set_title('beuler, t= '+str(tend/npts))
```

```python
stability(40, 30, 10000000, 30)
```

```python
stability(41, 30, 10000000, 30)
```

```python
stability(39, 100, 10000000, 30)
```
